```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
st = PorterStemmer()
stop = stopwords.words('english')
```

```python
def pre_processing(Data):
    Data = Data.apply(lambda x: " ".join(x.lower() for x in x.split())) #Convering to lower cases
    Data = Data.str.replace('[^\w\s]','') #Removing punctuations
    Data = Data.apply(lambda x: " ".join(x for x in x.split() if x not in stop)) # Removing stopwords
    Data = Data.apply(lambda x: " ".join([st.stem(word) for word in x.split()])) #Apply stemming
    return(Data) 
    
    
```

```python
data=pre_processing(Data_1['Review Text'])
```
