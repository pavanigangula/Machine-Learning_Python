```python
from sklearn.datasets import load_boston
import pandas as pd
boston = load_boston()
dataset = pd.DataFrame(data=boston.data, columns=boston.feature_names)
dataset['target'] = boston.target
print(dataset.head())
```

```python
X = dataset[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
Y = dataset['target']
```

```python
cormat = X.corr()
cormat
```

```python
import statsmodels.api as sm
# statsModel = sm.ols('Y~X',data = dataset)

#priceDF = pd.DataFrame(price, columns=list('x'))
#saleDF = pd.DataFrame(sale, columns=list('y'))

#import statsmodels.formula.api as smf
#fittedModel = smf.OLS(Y,X, data = dataset).fit()
X = sm.add_constant(X) # adding a constant

fittedModel = sm.OLS(Y, X).fit()
# fittedModel = sm.OLS(Y,X).fit()
```

```python
predictions = fittedModel.predict(X) 

print_model = fittedModel.summary()
print(print_model)
```
