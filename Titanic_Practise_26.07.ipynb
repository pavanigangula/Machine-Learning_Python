```python
def complete_data():
    train = pd.read_csv('C:\\Users\\Pavani Gangula\\Downloads\\titanic-train.csv')
    test = pd.read_csv("C:\\Users\\Pavani Gangula\\Downloads\\titanic-test.csv")
    targets =  train.Survived
    train.drop('Survived',1,inplace= True) # dropping 'loan_status' variable
    complete = train.append(test)             # combining train and test data
    complete.reset_index(inplace = True)
    complete.drop(['index','Cabin'],inplace=True, axis =1) # dropping loan-ID and index
    return complete
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
def compute_score(clf, X, y, scoring='accuracy'):
    xval = cross_val_score(clf, X, y, cv = 5, scoring=scoring)
    return np.mean(xval)
```

```python
def recover_train_test_target():
    global complete, train
    targets = train['Survived'].map({'Y':1,'N':0})
    train = complete.head(891)
    test = complete.iloc[891:]
    return train, test, targets
```
