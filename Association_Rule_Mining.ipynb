```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
whiskey = pd.read_csv("/home/Pavani Gangula/Desktop/Datasets/whiskey.csv")
```

```python
whiskey.head()
```

```python
freq_itemset = apriori(df=whiskey, min_support=0.1, use_colnames=True)
```

```python
rules = association_rules(freq_itemset, min_threshold=0.5)
```

```python
rules.sort_values(by='lift',ascending=False)
```

```python
store_data = pd.read_excel("/home/subhasis/Dropbox/Store data.xlsx")
```

```python
store_data.head()
```

```python
store_data['cnt'] = 1
```

```python
basket = store_data.groupby(['CASH_MEMO','PRODUCT'])['cnt'].sum().unstack().reset_index().fillna(0).set_index('CASH_MEMO')
```

```python
def recode(x):
    if x > 0:
        return 1
    else:
        return 0
basket = basket.applymap(recode)
```

```python
basket.head()
```

```python
freq_items_store = apriori(basket, use_colnames=True, min_support=0.005)
```

```python
rules_store = association_rules(freq_items_store, min_threshold=0.4)
```
